```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pydataset import data
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.heatmap(df.corr(),annot=True);
```

##### From this heatmap we can identify that totall_bill has a correlation with tips and size colums of the dataframe

```python
plt.axhline(y= df.tip.mean(),color='r')
sns.lineplot(y='tip',x='total_bill',data=df);
```

##### Here from this plot we can identify that there may be a hike or decrease the aqmount of tip with respect to totall bill upto a certain level. Aproximately from 45$ (total_bill) onwards the tip gets a huge hike

```python
sns.scatterplot(y='tip',x='total_bill',data=df);
```

##### this scatter plot shows a positive correlation.as tip increases with total_bill. 

```python
sns.jointplot(x='tip',y='total_bill',data=df);
```

##### both of the histogram shows a right skewed plot with also shows a correlation between them

### Regression plot

```python
sns.regplot(x='tip',y='total_bill',data=df);
```

##### This regression plot shows a positive correlation line between tip and total bill

```python
sns.barplot(x='time',y='tip',hue='sex', data=df);
```

##### from this graph we can identify that more people give tips at dinner time irrespective of gender

```python
sns.barplot(x='day',y='total_bill',hue='time',data=df);
```

##### This shows that on saturday and sunday there is no bill about lunch. so, might be the customers for lunch are working professionals. but, everyday from thursday to sunday the dinner bill amount gets increasing .There are some persons comes to visit that hotel or that area for some purpose. for having a delicious dinner, or for enjoy the night time
