### Univariate Analysis

```python
# 1) To find the popular cuisine based on age(40 plus)
Popular_Cuisine_Age=data.groupby(['Cuisines'])['Age'].value_counts()
Popular_Cuisine_Age
```

```python
sns.barplot(data=data,x='Cuisines',y='Age')
plt.show()
```

```python
# 2) To find the popular cuisine based on activity
sns.barplot(x='Cuisines', y='FoodRating', hue='Activity', data=data)
plt.xticks(rotation=60)
plt.title('Popular cuisine based on activity')
plt.show()
```

```python
# 3) To find the area with maximum number of restaurants
pd.crosstab(data.Location,data.Cuisines,margins=True)
```

```python
plt.xticks(rotation=60)
sns.countplot(data=data,x='Location',hue='Cuisines')
plt.show()
```

```python
plt.figure(figsize=(5,5))
x=data.Location.value_counts()
x.plot(kind='pie',fontsize=10)
plt.title('Number of restaurants in each area',fontsize=30,color='darkblue')
plt.show()
```

```python
# 4) To find oldest and youngest japanese cuisine lover
Japanese_cuisine_lovers=data.query("Cuisines=='Japanese'")
Japanese_cuisine_lovers.sort_values(by='Age',ascending=False)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Age',data=Japanese_cuisine_lovers,hue='FoodRating')
plt.show()
```

```python
sns.barplot(data=data,x='Cuisines',y='Age')
plt.show()
```

```python
plt.figure(figsize = (10, 5))
cor=data.corr()
sns.heatmap(cor, annot = True, cmap="YlGnBu")
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
# 5) To find popular cuisine based on budget
# 6) To find popular cuisine based on marital status
# 7) To find popular cuisine based on gender
plt.figure(figsize=(15,15))
plt.subplot(3, 1, 1)
sns.barplot(x='Cuisines', y='FoodRating', hue='Marital Status', data=data)
plt.xticks(rotation=60)
plt.title('Popular cuisine based on marital status')


plt.figure(figsize=(15,15))
plt.subplot(3, 1, 2)
sns.barplot(x='Cuisines', y='Budget', hue='Activity', data=data)
plt.xticks(rotation=60)
plt.title('Cuisines with their expenditure')
plt.show()


plt.figure(figsize=(15,15))
plt.subplot(3, 1, 3)
sns.barplot(x='Cuisines', y='FoodRating',hue='Gender', data=data)
plt.xticks(rotation=60)
plt.title('Popular cuisine based on gender')
plt.show()
```
