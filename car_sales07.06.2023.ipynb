  ###  Multivariate analysis

### Bar plot

#### To create the below shown bar chart we have to create the frequency table. Frequency table can be made using the value_counts() method.

```python
# Here in this graph, we are finding the count value by only giving x value
data.Manufacturer.value_counts().plot.bar(figsize=(6,2), cmap = 'turbo_r');
#Ford has more number of sales in this graph
```

```python
# Here in this graph, we are looking for first 20 low priced vechicles 
#by choosing x value as 'Manufacturer' and y value as 'Price_in_thousands' 
data.sort_values(by= 'Price_in_thousands')[:20].plot.bar('Manufacturer','Price_in_thousands', figsize=(6,3), cmap='rainbow');
```

### Area chart

```python
data.plot.area('Manufacturer','Price_in_thousands', figsize=(6,3), cmap='Pastel2');
```
