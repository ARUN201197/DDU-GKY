# Boarded from Southampton

```python
# Survived male passengers who boarded from Southampton exceeding 24 yrs
Boarded_Southampton=data.query("Boarded =='Southampton' & Age>24 & Sex=='male'  & Survived == 1.0")
```

```python
Boarded_Southampton
```

# Boarded from Cherbourg

```python
# Survived passengers who boarded from Cherbourg belongigng to age group [24-40]
boarded_Cherbourg=data.query("Boarded=='Cherbourg' & 24<=Age<=40 & Survived==1.0" )
```

```python
boarded_Cherbourg
```

# Boarded from Queenstown

```python
# Survived passengers who boarded from Queenstown having siblings or spouse
boarded_Queenstown=data.query("Boarded=='Queenstown' & Survived==1.0 & SibSp==True ")
```

```python
boarded_Queenstown
```
