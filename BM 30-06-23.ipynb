### <center> Boolean Masking

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

#### Show the use of boolean masking on Series objects

##### Example 1

```python
twopowers_sr = pd.Series([1,2,4,8,16,32,64,128,256,512,1024])
BM = [False,False,False,True,False,False,True,True,True,True,True]
```

```python
twopowers_sr[BM]
```

##### Example 2

```python
BM = twopowers_sr<100
twopowers_sr[BM]
```

```python
BM = twopowers_sr>=100
twopowers_sr[BM]
```
