# EDA on tips dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pydataset import data
import warnings
warnings.filterwarnings('ignore')
```

### Loading dataset from seaborn library

```python
data = sns.load_dataset('tips')
data.head(10)
```

### Univariate analysis

```python
data.columns
```

```python
to_list = ['total_bill', 'tip', 'size']
```

#### Histograms

```python
for i in to_list:
    sns.histplot(data[i], kde=True)
    plt.show()
```

```python
for i in to_list:
    sns.boxplot(y = data[i])
    plt.show()
```

#### Barplot

```python
data
```

```python
sex_sum = data.groupby(['sex'])['total_bill'].sum()
sex_mean = data.groupby(['sex'])['total_bill'].mean()
```

```python
sex_sum.plot.bar();
```

```python
sex_mean.plot.bar();
```

```python
sex_tip_sum = data.groupby(['sex'])['tip'].sum()
sex_tip_mean = data.groupby(['sex'])['tip'].mean()
```

```python
sex_tip_sum.plot.bar();
```

```python
sex_tip_mean.plot.bar();
```

```python
data
```

```python
smoker_tip_sum = data.groupby(['smoker'])['tip'].sum()
smoker_tip_mean = data.groupby(['smoker'])['tip'].mean()
```

```python
smoker_tip_sum.plot.pie(autopct = '%1.2f%%');
```

```python
smoker_tip_mean.plot.pie(autopct = '%1.2f%%');
```
