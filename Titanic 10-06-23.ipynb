## Cross Tabulation

##### In statistics,a contingency table(also known as a cross tabulation or crosstab) is a type of table in  a matrix format that displays the(multivariate) frequency distribution of the variables. They are heavily used in survey research,business intelligence,engineering,and scientific research. They provide a basic picture of the interraction between two variables and can help find interactions between them. the term contingency table was first used by Karl Pearson in "On the Theory of Contingency and its relation to Association and Normal Correlation " ,[1] part of the Drapers' Company Research Memoirs Biometric Series I Published in 1904

```python
pd.crosstab(data.Pclass,data.Embarked)
```

```python
pd.crosstab(data.Pclass,data.Embarked,margins=True)
```
